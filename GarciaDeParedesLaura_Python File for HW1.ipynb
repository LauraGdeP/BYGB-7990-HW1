```python
# Importing libraries
from IPython.display import display
import pandas as pd

# Setting option to display all rows in the dataframe
pd.set_option("display.max_rows", None)

# Saving student data in a dataframe
students_df = pd.read_csv("Students.txt", sep="\t")
```

```python
# Defining function to display all student records

def all_records():
    display(students_df)
    
```

```python
# Defining function to display students whose FIRST name begins with a certain string (case insensitive) 

def firstname_records():
    firstname = input("Please enter student FIRST NAME:       ").lower()
        
    fn = students_df["First"].str.lower().str.startswith(firstname, na = False)
    
    #Adding a check in case of invalid input by the user
    if not firstname.isalpha():
        print("Invalid Input")
    elif students_df[fn].empty == True:
        print("No results match your search criteria")
    else:
        display(students_df[fn])
        
```

```python
# Defining function to display students whose LAST name begins with a certain string (case insensitive) 

def lastname_records():
    lastname = input("Please enter student LAST NAME:       ").lower()
        
    ln = students_df["Last"].str.lower().str.startswith(lastname, na = False)

    #Adding a check in case of invalid input by the user
    if not lastname.isalpha():
        print("Invalid Input")
    elif students_df[ln].empty == True:
        print("No results match your search criteria")
    else:
        display(students_df[ln])
        
```

```python
# Defining function to display all records for students whose graduating year is a certain year

def gradyear_records():
    gradyear = int(input("Please enter student GRADUATION YEAR:      "))
    
    gy = students_df[students_df["GradYear"] == gradyear]
    
    #Adding a check in case of invalid input by the user
    if gy.empty == True:
        print("No results match your search criteria")
    else:
        display(gy)
        
```

```python
# Defining function to display a summary report of number and percent of students in each program, for students graduating on a certain year 

def summary_report():
    gradyear_summary = int(input("Please enter student GRADUATION YEAR for SUMMARY REPORT:    "))
    
    year = students_df[students_df["GradYear"] == gradyear_summary]
    
    summary = year.groupby(["DegreeProgram"]).size().reset_index(name="TotalStudents")
    
    summary["PercentOfStudents"] = (summary["TotalStudents"] / summary["TotalStudents"].sum()) * 100
    
    #Adding a check in case of invalid input by the user
    if summary.empty == True:
        print("No results match your search criteria")
    else:
        display(summary)
        
```

```python
# Calling functions to display records based on user command

command = input("Please enter which records would you like to retrieve:    ").lower()

command_list = ["all", "first", "last", "year", "summary"]

if command == "all":
    all_records()
elif command == "first":
    firstname_records()
elif command == "last":
    lastname_records()
elif command == "year":
    gradyear_records()
elif command == "summary":
    summary_report()
elif command not in command_list:
    print("Invalid Input")
    
```
